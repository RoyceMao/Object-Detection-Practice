```python
import numpy as np
import tensorflow as tf
sess = tf.Session()
```

Base_Anchors

```python
def Base_Anchor(base_size, ratios, scales):
    """
    生成3*3的9种基本anchors（有多少个scales，就有多少种面积，square = w * h = w^2 * ratios = base_size^2 * scales^2，所以w = 1/sqrt(ratios) * base_size * scales）
    base_size:
    ratios:
    scales:
    """
    # 
    ratios = tf.constant(ratios, dtype=tf.float32)
    scales = tf.constant(scales, dtype=tf.float32)
    #
    ratios = tf.expand_dims(ratios, axis=0)
    scales = tf.expand_dims(scales, axis=1)
    #
    w = 1.0/tf.sqrt(ratios) * base_size * scales
    h = tf.sqrt(ratios) * base_size * scales
    #
    w = tf.reshape(w, (-1, 1))
    h = tf.reshape(h, (-1, 1))
    
    return tf.concat([-0.5*w, -0.5*h, 0.5*w, 0.5*h], axis=1) # tf.stack([-0.5*w, -0.5*h, 0.5*w, 0.5*h], axis=1) 
```

```python
# 测试
"""
注：上述tf函数返回的w与h是2维的，tf.stack处理后会增加维度为3维tensor。所以直接用不改变维度的concat方法
"""
base_size = 128
ratios =[0.5, 1, 2]
scales = [1, 2, 4]
base_anchors = Base_Anchor(base_size, ratios, scales)
print(base_anchors.eval(session=sess))
```

All_Anchors

```python
def Shift_Anchors(shape, base_anchors, stride):
    """
    以feature map上所有锚点为基准点，生成所有的anchors
    shape:
    base_anchors:
    stride:
    """
    H, W = shape
    #
    center_x = (tf.cast(tf.range(W), dtype=tf.float32) + tf.constant(0.5, dtype=tf.float32)) * stride
    center_y = (tf.cast(tf.range(H), dtype=tf.float32) + tf.constant(0.5, dtype=tf.float32)) * stride
    # 
    center_x, center_y = tf.meshgrid(center_x, center_y)
    # 
    center_x = tf.reshape(center_x, [-1]) # 【[-1]参数说明，直接打平为1维】
    center_y = tf.reshape(center_y, [-1])
    #
    shifts = tf.stack([center_x, center_y, center_x, center_y], axis=1)
    #
    shifts = tf.expand_dims(shifts, axis=0)
    base_anchors = tf.expand_dims(base_anchors, axis=1)
    #
    all_anchors = shifts + base_anchors
    
    return tf.reshape(all_anchors, [-1, 4])
```

```python
# 测试
"""
注：上述函数最后的reshape操作，不能少，因为扩围后的两两操作返回的是对应的（a，b）维度，不是我们需要的打平的维度
"""
from skimage import io
from matplotlib import pyplot as plt
import cv2
img_path = r'C:\Users\Royce\Desktop\time.jpg' 
img = io.imread(img_path)
if img.ndim != 3:
    img = skimage.color.gray2rgb(img)
print('img size 大小为：{}'.format(img.shape))
# 得到all_anchors
stride = 20
feature_map_shape = (326, 550)
all_anchors = Shift_Anchors(feature_map_shape, base_anchors, stride)
print(all_anchors)
# 画图并可视化
for a in all_anchors.eval(session=sess):
    x1,y1,x2,y2 = a
    bbox = cv2.rectangle(img,(x1,y1),(x2,y2),(255, 165, 0),1)
plt.imshow(bbox)
```
