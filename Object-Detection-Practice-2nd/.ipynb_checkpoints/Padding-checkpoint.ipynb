```python
import numpy as np
import tensorflow as tf
sess = tf.Session()
```

加上padding

```python
def Padding_Pic(fixed_size, pic_tf):
    """
    给图片的宽/高，或某个维度上加上具体size的padding
    fixed_size:
    pic_np:3维
    """
    shape = tf.shape(pic_tf)
    # 增加1的标志位
    pic_tf_tag = tf.pad(pic_tf, [[0,0],[0,0],[0,1]], mode='CONSTANT', constant_values=1)
    # 在宽度/高度上均增加fixed_size的padding
    padding_size_H = tf.maximum(0, tf.cast((fixed_size-shape[0]) / 2, dtype=tf.int32))
    padding_size_W = tf.maximum(0, tf.cast((fixed_size-shape[1]) / 2, dtype=tf.int32))

    pic_tf_padding = tf.pad(pic_tf_tag, [[padding_size_H,padding_size_H],[0,0],[0,0]], mode='CONSTANT', constant_values=0)
    pic_tf_padding = tf.pad(pic_tf_padding, [[0,0],[padding_size_W,padding_size_W],[0,0]], mode='CONSTANT', constant_values=0)
    
    return pic_tf_padding
```

```python
# 测试
from PIL import Image
from skimage import io
import matplotlib.pyplot as plt
image_path = r'C:\Users\Royce\Desktop\time.jpg'
input_np = io.imread(image_path)
print("样例图片的shape为：{}".format(input_np.shape))
fixed_size = 580
print("固定尺寸的（580，580）的padding结果为：")
input_tf = tf.convert_to_tensor(input_np, dtype=tf.float32)
padded_result = Padding_Pic(fixed_size, input_tf)
padded_result_np = padded_result.eval(session=sess)
img = Image.fromarray(padded_result_np[...,:-1].astype('uint8'))
plt.imshow(img)
```

去除padding

```python
def Padding_remove(padded_result):
    """
    去除padding，宽、高两个维度都需要去padding
    padded_result
    """
    # 高度/宽度维度的tag求和项
    tag_sum_W = tf.reduce_sum(padded_result[...,-1], axis=0)
    tag_sum_H = tf.reduce_sum(padded_result[...,-1], axis=1)
    # 首先分别找到高度/宽度上的非padding索引范围
    start_W = tf.where(tag_sum_W > 0)[0][0]
    end_W = tf.where(tag_sum_W > 0)[-1][0]
    start_H = tf.where(tag_sum_H > 0)[0][0]
    end_H = tf.where(tag_sum_H > 0)[-1][0]
    
    return padded_result[start_H:end_H, start_W:end_W, :]
```

```python
# 测试
padded_remove_result = Padding_remove(padded_result)
padded_remove_result_np = padded_remove_result.eval(session=sess)
img = Image.fromarray(padded_remove_result_np[...,:-1].astype('uint8'))
plt.imshow(img)
```
