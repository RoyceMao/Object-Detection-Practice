```python
import numpy as np
import tensorflow as tf
sess = tf.Session()
```

回归目标

```python
def Regress_Target(gt_boxes, proposals):
    """
    回归目标
    gt_boxes:（num，4）
    proposals:（num，4）
    """
    # 中心点坐标
    ctr_x_gt = (gt_boxes[...,1] + gt_boxes[...,3]) / 2
    ctr_x_pp = (proposals[...,1] + proposals[...,3]) / 2
    ctr_y_gt = (gt_boxes[...,0] + gt_boxes[...,2]) / 2
    ctr_y_pp = (proposals[...,0] + proposals[...,2]) / 2
    # 宽度与高度
    width_gt = gt_boxes[...,3] - gt_boxes[...,1]
    width_pp = proposals[...,3] - proposals[...,1]
    height_gt = gt_boxes[...,2] - gt_boxes[...,0]
    height_pp = proposals[...,2] - proposals[...,0]
    # 计算回归目标
    deltas_x = (ctr_x_gt - ctr_x_pp) / width_pp
    deltas_y = (ctr_y_gt - ctr_y_pp) / height_pp
    deltas_width = tf.log(width_gt / width_pp)
    deltas_height = tf.log(height_gt / height_pp)
    # 转置
    deltas = tf.stack([deltas_x, deltas_y, deltas_width, deltas_height], axis=1)
    # 标准化
    deltas /= tf.constant([0.1, 0.1, 0.2, 0.2], dtype=tf.float32)
    
    return deltas
```

```python
# 测试
gt_boxes = np.array([[22,78,304,253],[19,269,296,360]])
anchors = np.array([[15,60,280,260],[50,269,296,360]])
gt_boxes = tf.constant(gt_boxes, dtype=tf.float32)
anchors = tf.constant(anchors, dtype=tf.float32)
deltas = Regress_Target(anchors, gt_boxes)
print("回归目标：\n{}".format(deltas.eval(session=sess)))
```

回归应用

```python
def Regress_Apply(proposals, deltas):
    """
    回归应用
    proposals:
    deltas:
    """
    # 中心点坐标
    ctr_x_pp = (proposals[...,1] + proposals[...,3]) / 2
    ctr_y_pp = (proposals[...,0] + proposals[...,2]) / 2
    # 宽度与高度
    width_pp = proposals[...,3] - proposals[...,1]
    height_pp = proposals[...,2] - proposals[...,0]
    # 标准化还原
    deltas *= tf.constant([0.1, 0.1, 0.2, 0.2], dtype=tf.float32)
    # 应用
    ctr_x_pp += deltas[...,0] * width_pp
    ctr_y_pp += deltas[...,1] * height_pp
    width_pp *= tf.exp(deltas[...,2])
    height_pp *= tf.exp(deltas[...,3])
    # 转化为（y1，x1，y2，x2）
    target = tf.stack([ctr_y_pp-height_pp*0.5, ctr_x_pp-width_pp*0.5, ctr_y_pp+height_pp*0.5, ctr_x_pp+width_pp*0.5], axis=1)
    
    return target
```

```python
# 测试
from skimage import io
import cv2
from matplotlib import pyplot as plt 
target = Regress_Apply(anchors, deltas)
print('回归修正坐标值为:\n{}'.format(sess.run(target)))

## 可视化
img_path = r'C:\Users\Royce\Desktop\time.jpg'
img_new = io.imread(img_path)
target_np = target.eval(session=sess)
# gt_boxes = gt_boxes.eval(session=sess)
for a in target_np:
    y1,x1,y2,x2 = a
    bbox = cv2.rectangle(img_new,(x1,y1),(x2,y2),(255, 165, 0),10) # 橙色的修正anchor
for a in gt_boxes.eval(session=sess):
    y1,x1,y2,x2 = a
    bbox = cv2.rectangle(img_new,(x1,y1),(x2,y2),(0, 165, 0),3) # 绿色gt
plt.imshow(img_new)
```
