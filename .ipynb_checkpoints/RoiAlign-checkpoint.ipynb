```python
import numpy as np
import tensorflow as tf
```

```python
sess = tf.Session()
```

```python
# 简单的类VGG 8倍下采样
from __future__ import print_function
from __future__ import absolute_import

from keras.models import Model
from keras.layers import Conv2D, Reshape, Input, Activation, Convolution2D, MaxPooling2D, ZeroPadding2D, Add, BatchNormalization, concatenate, TimeDistributed, AveragePooling2D, Flatten, Dense
from keras import backend as K

def vgg_like(input_tensor):
    """
    VGG网络的前3个blocks，8倍下采样
    :param input_tensor: 
    :param trainable: 
    :return: 
    """
    bn_axis = 3

    # Block 1
    x = Conv2D(64, (3, 3), activation='relu', padding='same', name='block1_conv1')(input_tensor)
    x = Conv2D(64, (3, 3), activation='relu', padding='same', name='block1_conv2')(x)
    x = MaxPooling2D((2, 2), strides=(2, 2), name='block1_pool')(x)

    # Block 2
    x = Conv2D(128, (3, 3), activation='relu', padding='same', name='block2_conv1')(x)
    x = Conv2D(128, (3, 3), activation='relu', padding='same', name='block2_conv2')(x)
    x = MaxPooling2D((2, 2), strides=(2, 2), name='block2_pool')(x)

    # Block 3
    x = Conv2D(256, (3, 3), activation='relu', padding='same', name='block3_conv1')(x)
    x = Conv2D(256, (3, 3), activation='relu', padding='same', name='block3_conv2')(x)
    x = Conv2D(256, (3, 3), activation='relu', padding='same', name='block3_conv3')(x)
    fp = MaxPooling2D((2, 2), strides=(2, 2), name='block3_pool')(x)
#     tf.add_to_collection('activations', fp) # 可视化feature map
    
    return x
```

```python
from skimage import io
import matplotlib.pyplot as plt
img_path = r'C:\Users\Royce\Desktop\EG.png' 
img = io.imread(img_path)
img_batch = np.expand_dims(img, axis=0)
print(img_batch.shape)
img_tensor = tf.constant(img_batch[...,:3], dtype=tf.float32)
input_tensor = img_tensor
feature_map = vgg_like(input_tensor)
# visualize_layers = ['fp']
# conv_out = sess.run(tf.get_collection('activations'), feed_dict={x: x_train_a, y_: y_train_a})
print(feature_map.shape)
```

样例之：原图上做RoiAlign

```python
# feature map暂未可视化成功，先在原图上做RoiAlign
# tf.image.crop_and_resize()方法
'''
tf.image.crop_and_resize(
    image,:图片tensor
    boxes,:bbox框tensor，必须是 [y1,x1,y2,x2]的顺序
    box_ind,:
    crop_size,:
    method='bilinear',:
    extrapolation_value=0,:
    name=None
)
image：    一个Tensor,必须是下列类型之一：uint8,uint16,int8,int16,int32,int64,half,float32,float64,一个形状为[batch, image_height, image_width, depth]的四维张量,image_height和image_width需要为正值.
boxes：    一个类型为float32的Tensor,是形状为[num_boxes, 4]的二维张量.张量的第i行指定box_ind[i]图像中框的坐标,并且在标准化坐标中指定[y1, x1, y2, x2]；标准化的坐标值y被映射到图像坐标y * (image_height - 1)处,从而标准化图像高度的[0, 1]间隔被映射到[0, image_height - 1]的图像高度坐标中.我们允许y1> y2,在这种情况下,采样的裁剪是原始图像的上下翻转版本.宽度维度的处理方式类似.[0, 1]范围之外的标准化坐标是允许的,在这种情况下,我们使用extrapolation_value外推输入图像值.
box_ind：  一个int32类型的Tensor；形状为[num_boxes]的1维张量,在[0, batch)中具有int32值.该box_ind[i]值指定第i个方框要引用的图像.
crop_size：一个int32类型的Tensor；一个2个元素的一维张量,size = [crop_height, crop_width].所有裁剪的图像修补程序都调整为此大小.图像内容的宽高比不被保留；crop_height和crop_width需要为正值.
method：可选的string,其来自："bilinear"；默认为"bilinear"；指定插值方法的字符串.现在只支持“双线性(bilinear)”.
。。。
'''
fp_tensor = tf.convert_to_tensor(feature_map, dtype=tf.float32)
# 参数1
im_tensor = tf.expand_dims(img,axis=0) # 直接增加1维batch
# print(im_tensor)
sample_boxes = tf.constant([[190,75,240,120],[190,325,240,280]],dtype=tf.float32)
h =  300
w =  400
y1 = sample_boxes[...,0]/h
x1 = sample_boxes[...,1]/w
y2 = sample_boxes[...,2]/h
x2 = sample_boxes[...,3]/w
# 参数2
sample_boxes_normal = tf.stack([y1,x1,y2,x2],axis = 1)
# print(sample_boxes_normal)
# 参数3、4
box_ind = tf.constant([0,0],dtype=tf.int32)
crop_size = tf.constant([200,200],dtype=tf.int32)
# 开始RoiAlign
crop_tensors = tf.image.crop_and_resize(im_tensor,
                                        sample_boxes_normal,
                                        box_ind,
                                        crop_size,
                                        method='bilinear'
                                        )
print(crop_tensors)
# 可视化crop出来的tensor图片
crop_nps = crop_tensors.eval(session=sess)
print(crop_nps.shape)
for i,crop_np in enumerate(crop_nps):
    im=Image.fromarray(crop_np.astype('uint8')) # 可视化feature map
    plt.subplot(1,2,i+1)
    plt.imshow(im)
```

样例之：在feature_map上做RoiAlign

```python
# 继续在feature_map上做RoiAlign
img_path = r'C:\Users\Royce\Desktop\time.jpg'
img = io.imread(img_path)
# 增加batch维度
img_batch = np.expand_dims(img, axis=0)
print(img_batch.shape)
img_tensor = tf.constant(img_batch[...,:3], dtype=tf.float32)
feature_map = vgg_like(img_tensor)
print(feature_map)
```

```python
# 直接在feature map上做RoiAlign
crop_feature_tensors = tf.image.crop_and_resize(feature_map,
                                        sample_boxes_normal,
                                        box_ind,
                                        crop_size,
                                        method='bilinear'
                                        )
```

```python
# 可视化crop出来的feature_tensor
#------------------------------------- 这两句做参数初始化的操作，必须有，不然下一句tensor转numpy会报错
init = tf.global_variables_initializer()
sess.run(init) 
#-------------------------------------
crop_feature_nps = crop_feature_tensors.eval(session=sess)
print(crop_feature_nps.shape)
for i,crop_np in enumerate(crop_feature_nps):
    print(crop_np.shape)
    im=Image.fromarray(crop_np[...,:3].astype('uint8')) # 可视化feature_map，只选取前3个channel
    plt.subplot(1,2,i+1)
    plt.imshow(im)
```
